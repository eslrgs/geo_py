### 5. Loops

```python
numbers = [1, 2, 3, 4, 39, 50, 700]

for number in numbers:
    
    if number < 40:
    
        print(number)
```

```python
euan = True

while euan == True:
    
    print('this will break my computer')
```

```python
x = 1

while x < 20:
    
    print(x)
    
    x += 1
```

### 7. Classes and methods (functions)

```python
class my_maths:
    
    def add(x, y):
    
        return x + y
    
    def subtract(x, y):

        return x - y

print(my_maths.add(2, 3))
print(my_maths.subtract(2, 3))

m = my_maths

m.add(100, 4)
```
