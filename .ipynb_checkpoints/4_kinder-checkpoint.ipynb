```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
import matplotlib as mp
import mplstereonet as mpl
from shapely.geometry import Point

%config InlineBackend.figure_format='retina'
```

```python
geo_map = gpd.read_file('shp/kinder_map.shp')
geo_map.to_crs(epsg = 27700, inplace = True)
kinder = geo_map[geo_map.LEX_D.str.contains('KINDERSCOUT')]

beds = pd.read_csv('data/upper_beds.csv')

def prep(data):

    data.columns = data.columns.str.lstrip()
    geometry = [Point(xy) for xy in zip(data.longitude, data.latitude)]

    data['geometry'] = geometry
    data = data.set_geometry('geometry')
    data.crs = {'init': 'epsg:4326'}
    data.to_crs(epsg = 27700, inplace = True)
    data['dist'] = np.cumsum(data.distance(data.shift()))
    
    return data

df = prep(beds)
```

```python
first = df[df.dist < 1200]
second = df[df.dist >= 1200]

fig, ax = mpl.subplots(figsize = (9, 6))
ax.grid(color = 'k', alpha = 0.2)

strike1, dip1, azi1 = first.strike, first.dip, first.dipAzimuth
strike2, dip2, azi2 = second.strike, second.dip, second.dipAzimuth

ax.plane(strike1, dip1, color = 'tab:red', linewidth = 2, alpha = 0.5)
ax.pole(strike1, dip1, color = 'tab:red',  marker = 'o', markersize = 5, alpha = 1)
note = f"0 - 1200 m\nN = {dip1.size}\nMean dip: {azi1.mean():.0f}\nMean strike: {strike1.mean():.0f}"
ax.annotate(note, xy = (5 * 72, 45), color = 'tab:red', xycoords='axes points')

ax.plane(strike2, dip2, color = 'tab:blue', linewidth = 2, alpha = 0.5)
ax.pole(strike2, dip2, color = 'tab:blue',  marker = 'o', markersize = 5, alpha = 1)
note = f"1200 - 1750 m\nN = {dip2.size}\nMean dip: {azi2.mean():.0f}\nMean strike: {strike2.mean():.0f}"
ax.annotate(note, xy = (5 * 72, 110), color = 'tab:blue', xycoords='axes points')
```

```python
fig, ax = plt.subplots(figsize = (15, 6))

kinder.plot(alpha = 0.5, color = 'tab:orange', ax = ax)

for index, row in first.iterrows():

    angle = row.strike
    t = mp.markers.MarkerStyle(marker='|')
    
    t._transform = t.get_transform().rotate_deg(angle*-1)
    first.loc[[index]].plot(c = 'tab:red', marker = t, markersize = 300, ax = ax)
    
for index, row in second.iterrows():

    angle = row.strike
    t = mp.markers.MarkerStyle(marker='|')
    
    t._transform = t.get_transform().rotate_deg(angle*-1)
    second.loc[[index]].plot(color = 'tab:blue', marker = t, markersize = 300, ax = ax)
    
#     angle = row.dipAzimuth
#     t._transform = t.get_transform().rotate_deg(angle*-1)
#     up.loc[[index]].plot(color = 'k', marker = t, markersize = 25, ax = ax)

ax.set(xlim = (df.geometry.x.min() - 200, df.geometry.x.max() + 300),
               ylim = (df.geometry.y.min() - 100, df.geometry.y.max() + 100))
```
