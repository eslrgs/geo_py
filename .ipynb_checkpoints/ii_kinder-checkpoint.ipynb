```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
import fiona
import matplotlib as mp
import contextily as ctx
import mplstereonet as mpl
import matplotlib.colors as mcolors
import matplotlib.cm as cm

from scipy import stats
from scipy.signal import savgol_filter
from shapely.geometry import Point

plt.rcParams.update({'font.size': 12})
plt.rcParams['font.family'] = "Arial"

%config InlineBackend.figure_format='retina'
```

```python
geo_map = gpd.read_file('/Users/euansoutter/Dropbox (Personal)/python/peak/shp/map_merge.shp')
geo = geo_map[geo_map.LEX_D.str.contains('KINDERSCOUT')]
geo.to_crs(epsg = 27700, inplace = True)

l = pd.read_csv('data/lower_beds.csv')
h = pd.read_csv('data/upper_beds.csv')

def prep(data):

    data.columns = data.columns.str.lstrip()
    geometry = [Point(xy) for xy in zip(data.longitude, data.latitude)]

    data['geometry'] = geometry
    data = data.set_geometry('geometry')
    data.crs = {'init': 'epsg:4326'}
    data.to_crs(epsg = 27700, inplace = True)
    data['dist'] = np.cumsum(data.distance(data.shift()))
    
    return data

low = prep(l)
high = prep(h)
```

```python
low.replace(' Top', 3, inplace = True)
low.replace(' Mid', 2, inplace = True)
low.replace(' Bottom', 1, inplace = True)
low.replace(' Channelly', 0, inplace = True)
low.replace(' Scours', 0, inplace = True)
high.dist.replace(np.nan, 0, inplace = True)
low.dist.replace(np.nan, 0, inplace = True)
```

```python
fig, ax = plt.subplots(figsize = (8, 6), nrows = 2)

def plots(data, var, ax):
    
    ax = ax
    ax.plot(data.dist, data[var], 'o', c = 'k', markersize = 1)
#     ax.plot(data.dist, savgol_filter(data[var], 5, 3), c = 'grey', lw = 0.5)
    ax.plot(data.dist, savgol_filter(data[var], 11, 3), c = 'k', lw = 2)
#     print(stats.spearmanr(data.dist, savgol_filter(data[var], 11, 3)))
#     print(stats.spearmanr(data.dist, data[var]))

plots(high, 'dipAzimuth', ax[0])
ax[0].set(ylim = (200, 290), xlim = (0, 1800),
          ylabel = 'dip direction [°]', xlabel = 'distance along outcrop [m]')

plots(high, 'dip', ax[1])
ax[1].set(ylim = (10, 30), xlim = (0, 1800),
          ylabel = 'dip [°]', xlabel = 'distance along outcrop [m]')

plt.tight_layout()

plt.savefig('figs/plot.jpg', dpi = 300)
```

```python
first = high[high.dist < 1200]
second = high[high.dist >= 1200]

fig, ax = mpl.subplots(figsize = (9, 6))
ax.grid(color = 'k', alpha = 0.2)

strike1, dip1, azi1 = first.strike, first.dip, first.dipAzimuth
strike2, dip2, azi2 = second.strike, second.dip, second.dipAzimuth

ax.plane(strike1, dip1, color = 'tab:red', linewidth = 2, alpha = 0.5)
ax.pole(strike1, dip1, color = 'tab:red',  marker = 'o', markersize = 5, alpha = 1)
note = f"0 - 1200 m\nN = {dip1.size}\nMean dip: {azi1.mean():.0f}\nMean strike: {strike1.mean():.0f}"
ax.annotate(note, xy = (5 * 72, 45), color = 'tab:red', xycoords='axes points')

ax.plane(strike2, dip2, color = 'tab:blue', linewidth = 2, alpha = 0.5)
ax.pole(strike2, dip2, color = 'tab:blue',  marker = 'o', markersize = 5, alpha = 1)
note = f"1200 - 1750 m\nN = {dip2.size}\nMean dip: {azi2.mean():.0f}\nMean strike: {strike2.mean():.0f}"
ax.annotate(note, xy = (5 * 72, 110), color = 'tab:blue', xycoords='axes points')

plt.savefig('figs/stereo.jpg', dpi = 300)
```
