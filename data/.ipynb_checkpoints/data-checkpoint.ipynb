```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(columns = ['height', 'bed_number', 'bed_thickness', 'grainsize', 'paleocurrent', 'strike', 'dip'])
```

```python
N = 100
```

```python
df['bed_number'] = np.linspace(1, 101, N)
```

```python
thickness = np.linspace(2, 5, 50) + np.random.normal(0, 1.5, 50)
thickness = np.append(thickness, np.linspace(20, 40, 50) + np.random.normal(0, 5, 50))
df['bed_thickness'] = thickness
```

```python
height = df.bed_thickness.cumsum() + np.random.normal(0, 5, 100)
df['height'] = height
```

```python
df
```

```python
grainsize = np.linspace(0.5, 0.5, 5)
grainsize = np.append(grainsize, np.linspace(0.5, 2, 45) + np.random.normal(0, 0.5, 45))
grainsize = np.append(grainsize, np.linspace(4, 5, 50) + np.random.normal(0, 0.5, 50))
df['grainsize'] = grainsize
df['grainsize'] = df['grainsize'].round(0)

# gsize = ['mud', 'silt', 'f. sand', 'm. sand', 'c.sand', 'congl.']

# df['grainsize'] = df['grainsize'].replace(0., gsize[0])
# df['grainsize'] = df['grainsize'].replace(1., gsize[1])
# df['grainsize'] = df['grainsize'].replace(2., gsize[2])
# df['grainsize'] = df['grainsize'].replace(3., gsize[3])
# df['grainsize'] = df['grainsize'].replace(4., gsize[4])
# df['grainsize'] = df['grainsize'].replace(5., gsize[4])
# df['grainsize'] = df['grainsize'].replace(6., gsize[5])

df
```

```python
fm = np.repeat('Liverpool Formation', 80)
fm = np.append(fm, np.repeat('Manchester Formation', 20))
df['formation'] = fm
```

```python
x = np.linspace(310, 10, 50) + np.random.normal(0, 2, 50)
x1 = np.linspace(360, 355, 50) + np.random.normal(0, 2, 50)
df['paleocurrent'] = np.concatenate((x, x1))
```

```python
x = np.random.uniform(low = 220, high = 240, size = 100) + np.random.normal(0, 10, 100)
df['strike'] = x
df['strike'].iloc[::2] = None
```

```python
x = np.linspace(40, 35, 50) + np.random.normal(0, 2, 50)
x1 = np.linspace(35, 30, 50) + np.random.normal(0, 2, 50)
df['dip'] = np.concatenate((x, x1))
df['dip'].iloc[::2] = None
```

```python
d18O = np.geomspace(0.2, 0.6, 50) + np.random.normal(0, 0.1, 50)
d18O = np.append(d18O, np.geomspace(1.3, 1., 50) + np.random.normal(0, 0.1, 50))
df['d18O'] = d18O
```

```python
cols = ['grainsize', 'paleocurrent', 'strike', 'dip']
cols1 = ['height', 'bed_number', 'bed_thickness', 'd18O']

df[cols] = df[cols].round(0)
df[cols1] = df[cols1].round(2)
```

```python
df.to_excel('data.xlsx')  
```
